```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['promised_delivery_time'] = pd.to_datetime(df['promised_delivery_time'])
df['actual_delivery_time'] = pd.to_datetime(df['actual_delivery_time'])
df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
plt.hist(df['order_total'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Order Total')
plt.xlabel('Total Order (₹)')
plt.ylabel('Number of Orders')
plt.grid(True)
plt.savefig("1.png")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (6,4))
payment_counts=df['payment_method'].value_counts()
plt.bar(payment_counts.index, payment_counts.values, color='green')
plt.title('Payment Method Distribution')
plt.xlabel('Payment Method')
plt.ylabel('Number of Orders')
plt.savefig("2.jpg")
plt.tight_layout()
plt.show()
```

```python
df['month'] = df['order_date'].dt.to_period('M')
monthly_orders = df['month'].value_counts().sort_index()
plt.figure(figsize = (10,5))
plt.plot(monthly_orders.index.astype(str), monthly_orders.values, marker='o', color='black')
plt.title('Number of Orders per Month')
plt.xlabel('Month')
plt.ylabel('Orders')
plt.xticks(rotation=45)
plt.savefig("3.png")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5, 5))
status_counts = df['delivery_status'].value_counts()
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Delivery Status Distribution')
plt.tight_layout()
plt.savefig("4.png")
plt.show()
```

```python
df['delivery_diff_min'] = (df['actual_delivery_time'] - df['promised_delivery_time']).dt.total_seconds() / 60
plt.figure(figsize=(8, 4))
plt.hist(df['delivery_diff_min'], bins=30, color='purple', edgecolor='black')
plt.title('Delivery Time Difference (Actual - Promised)')
plt.xlabel('Time Difference (minutes)')
plt.ylabel('Number of Orders')
plt.savefig("5.png")
plt.tight_layout()
plt.show()
```

```python
top_stores = df['store_id'].value_counts().head(10)
plt.figure(figsize=(8, 5))
plt.bar(top_stores.index.astype(str), top_stores.values, color='teal')
plt.title('Top 10 Stores by Number of Orders')
plt.xlabel('Store ID')
plt.savefig("6.jpg")
plt.ylabel('Number of Orders')
plt.tight_layout()
plt.show()
```

```python
top_partners = df['delivery_partner_id'].value_counts().head(10)
plt.figure(figsize=(8, 5))
plt.barh(top_partners.index.astype(str), top_partners.values, color='coral')
plt.title('Top 10 Delivery Partners by Order Count')
plt.xlabel('Number of Orders')
plt.savefig("8.png")
plt.ylabel('Delivery Partner ID')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7, 5))
plt.scatter(df['order_total'], df['delivery_diff_min'], alpha=0.6, color='red')
plt.title('Order Total vs. Delivery Time Difference')
plt.xlabel('Total Order(₹)')
plt.ylabel('Time Difference (minutes)')
plt.grid(True)
plt.savefig("9.jpg")
plt.tight_layout()
plt.show()
```

```python
payment_status = pd.crosstab(df['payment_method'], df['delivery_status'])
payment_status.plot(kind='bar', stacked=True, figsize=(8, 5), colormap='tab20')
plt.title('Payment Method by Delivery Status')
plt.xlabel('Payment Method')
plt.ylabel('Number of Orders')
plt.savefig("10.png")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (8,5))
sns.histplot(data = df,x='order_total',bins = 30,kde=True,color = 'skyblue')
plt.title("Distribution of order Totals")
plt.xlabel("Order Total")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='payment_method', y='order_total', palette='Set2')
plt.title("Order Total by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Order Total")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='delivery_status', palette='Set1')
plt.title("Delivery Status Count")
plt.xlabel("Delivery Status")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
daily_orders = df['order_date'].dt.date.value_counts().sort_index()
plt.figure(figsize=(12, 5))
sns.lineplot(x=daily_orders.index, y=daily_orders.values, marker="o", color="green")
plt.title("Daily Order Count Over Time")
plt.xlabel("Order Date")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
